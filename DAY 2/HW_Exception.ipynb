```python

import os

class CustomeExceptionForEmptyFile(Exception):
    pass
class CustomeExceptionForBigFile(Exception):
    pass

def read_file(file_path):
    try:
        file = open(file_path,"r")
        
        if os.path.getsize(file_path) == 0:
            raise CustomeExceptionForEmptyFile("THE FILE IS EMPTY YOU MUSTT READ UNEMPTY FILE")
        
        if os.path.getsize(file_path) > 10000000:
            raise CustomeExceptionForBigFile("THIS FILE IS BIGGER THAN 10 mb YOU MUSTT READ SMALLER FILE")
        
        
        print("if you print this line no exception was ras congratulations   ")
        print("WOOHOOOOO !!")
        file.close()
    except FileNotFoundError as ex: 
        print(ex)  
    except CustomeExceptionForEmptyFile as ex:
        print(ex) 
    except CustomeExceptionForBigFile as ex:
        print(ex) 
    except Exception as ex:
        print(ex)
```
